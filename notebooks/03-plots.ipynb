```python
anscombe = sns.load_dataset('anscombe')
```

```python
anscombe
```

```python
d1 = anscombe[anscombe['dataset'] == 'I']
```

```python
plt.plot(d1['x'], d1['y'], 'o')
```

```python
d2 = anscombe[anscombe['dataset'] == 'II']
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.plot(d1['x'], d1['y'], 'o')
ax2.plot(d2['x'], d2['y'], 'o')
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,
           fit_reg=False,
           hue='sex', col='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,
           fit_reg=False,
           hue='sex', col='day', row='smoker')
```

```python
sns.distplot(tips['tip'])
plt.show()
```
